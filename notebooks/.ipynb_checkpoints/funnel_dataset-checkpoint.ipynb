```python
import pandas as pd

df = pd.read_csv("/Users/prerana/Documents/Marketing_Conversion_Analytics/data/marketing_conversion.csv")

funnel = pd.DataFrame({
    "Stage": [
        "Email Opens",
        "Email Clicks",
        "Website Visits",
        "Conversions"
    ],
    "Count": [
        df["EmailOpens"].sum(),
        df["EmailClicks"].sum(),
        df["WebsiteVisits"].sum(),
        df["Conversion"].sum()
    ]
})

funnel
```

```python
funnel.to_csv(
    "funnel_data.csv",
    index=False
)
```
